```python
import numpy as np
import cv2
import os
import pandas as pd
import glob
```

```python
path='/home/ubuntu/gcubme4/Workspace/YJ_Kim/resnet50/rawdata_edit/abnormal/*'
paths=glob.glob(path)
```

```python
for i in paths[12:]:
    img_file='/home/ubuntu/gcubme4/Workspace/YJ_Kim/resnet50/rawdata_edit/abnormal/'
    fname = i[i.rindex('/')+1:]
    img_path=os.path.join(img_file, fname)
    print(img_path)
    
    img = cv2.imread(img_path)
    img_yuv=cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
    
    img_clahe = img_yuv.copy()
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8)) #CLAHE 생성
    img_clahe[:,:,0] = clahe.apply(img_clahe[:,:,0])           #CLAHE 적용
    img_clahe = cv2.cvtColor(img_clahe, cv2.COLOR_YUV2BGR)
    
    img_save_path ='/home/ubuntu/gcubme4/Workspace/YJ_Kim/exp_CLAHE/rawdata_edit/abnormal/'
    img_name = fname[:-16]
    cv2.imwrite(img_save_path+ img_name+'.jpg',img_clahe)
```
